```python
# ============================================
#   PCA & LDA Dimensionality Reduction to 2D
#   Dataset: Wine (included inside sklearn)
# ============================================

import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# --------------------------------------------
# 1. Load Wine Dataset
# --------------------------------------------
data = load_wine()
X = data.data
y = data.target
target_names = data.target_names

# --------------------------------------------
# 2. PCA: Reduce dimensions to 2 components
# --------------------------------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# --------------------------------------------
# 3. LDA: Reduce dimensions to 2 components
# --------------------------------------------
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X, y)

# --------------------------------------------
# 4. Visualization (2D scatter plots)
# --------------------------------------------
plt.figure(figsize=(12, 5))

# ----- PCA Plot -----
plt.subplot(1, 2, 1)
for i, name in enumerate(target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=name)
plt.title("PCA (Wine Dataset) – Reduced to 2 Dimensions")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()

# ----- LDA Plot -----
plt.subplot(1, 2, 2)
for i, name in enumerate(target_names):
    plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1], label=name)
plt.title("LDA (Wine Dataset) – Reduced to 2 Dimensions")
plt.xlabel("LD1")
plt.ylabel("LD2")
plt.legend()

plt.tight_layout()
plt.show()
```
